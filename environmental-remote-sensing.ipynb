```python
import pandas as pd
```

```python
soil_moisture_df = pd.read_csv(
    "soilmoisture_mozambique.csv", index_col="datetime", parse_dates=True
)
soil_moisture_df = soil_moisture_df[soil_moisture_df.location == "Mabalane"]
soil_moisture_df.index
```

```python
soil_moisture_df.groupby("type").plot(subplots=True)
```

```python
soil_moisture_df.groupby("type")["value"].plot(x="datetime", y="value", legend=True)
```
